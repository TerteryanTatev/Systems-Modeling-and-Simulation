```python
import numpy as np

A = np.array([
    [6, -3, 5],
    [1,  3, 5],
    [-2, 4, -3]
], dtype=float)

```

```python

y = np.array([1, 1, 1], dtype=float)


for k in range(50):

    y = A @ y
    norm_y = np.linalg.norm(y)
    x = y / norm_y
    lambda_vec = (A @ x) / x
    lambda_k = np.mean(lambda_vec)

    print(f"{k:1d} | {y.round(3)} | {x.round(3)} | {lambda_vec.round(3)} | {lambda_k:.3f}")
```

```python


y0 = np.array([1, 1, 1], dtype=float)

y_k_minus_1 = None
y_k = y0.copy()
y_k_plus1 = y0.copy()

lambda1_values = []
lambda2_values = []

n = 50
tarb = 0.00000001

for k in range(1, n):

    y_new = A @ y_k_plus1
    lam = y_new / y_k_plus1
    lambda1 = np.mean(lam)
    lambda1_values.append(lambda1)

    if y_k_minus_1 is not None:
        num = y_new - lambda1 * y_k_plus1
        den = y_k_plus1 - lambda1 * y_k_minus_1

        lam2 = num / den
        lambda2 = np.mean(lam2)
        lambda2_values.append(lambda2)

    y_norm = y_new / np.sqrt(np.sum(y_new**2))

    y_k_minus_1 = y_k.copy()
    y_k = y_k_plus1.copy()
    y_k_plus1 = y_norm.copy()

    if k > 2 and abs(lambda1_values[-1] - lambda1_values[-2]) < tarb:
        break

lambda1_final = lambda1_values[-1]
lambda2_final = lambda2_values[-1]
x = y_k_plus1

print("λ1 =", lambda1_final)
print("λ2 =", lambda2_final)
print("X1 =", x)
```
