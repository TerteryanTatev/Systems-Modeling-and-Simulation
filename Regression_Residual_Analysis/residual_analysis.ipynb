```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'N': [ i for i in range(1,16)],
    'x' :[4.7, 2.3, 3.0, 1.5, 3.1, 2.6, 1.0, 1.7, 3.6, 1.6, 4.3, 1.9, 2.0, 5.0, 3.1],
    'z' :[1.8, 2.6, 3.5, 2.1, 3.5, 4.0, 2.0, 4.5, 4.9, 5.2, 2.5, 4.6, 1.4, 0.4, 2.4]
})

df
```

```python

x = df['x'].values
z = df['z'].values
n = len(x)
Sxx = Sxz_anc = Szz_anc = Sxz = Szz = 0
for i in range(1, n):


        Sxx += x[i - 1] **2
        Sxz_anc += x[i - 1] * z[i - 1]
        Szz_anc += z[i - 1] **2
        Sxz += x[i - 1] * z[i]
        Szz += z[i - 1] * z[i]

A1 = np.array([[Sxx, Sxz_anc],
                  [Sxz_anc, Szz_anc]])
B1 = np.array([Sxz, Szz])
print('A1=',A1,'\n','B1=', B1)
a1, b1 = np.linalg.solve(A1, B1)

hn_z1 = np.zeros(n - 1)
SSE = 0.0

for i in range(1, n):
        hn_z1[i - 1] = a1 * x[i - 1] + b1 * z[i - 1]

        SSE += (z[i] - hn_z1[i - 1]) ** 2

sigma1 = np.sqrt(SSE / (n - 3))


print(f'a1= {a1}, b1= {b1}, sigma1 = {sigma1}')
Z1 = pd.DataFrame({
   ' z1': hn_z1
})

Z1


```
