```python

def get_R(columns):
    X = df[columns].values
    y = df['Y'].values


    X = np.column_stack((np.ones(len(X)), X))
    beta = np.linalg.inv(X.T @ X) @ (X.T @ y)
    y_hat = X @ beta
    y_mean = y.mean()


    SSR = ((y_hat - y_mean)**2).sum()
    SS0 = ((y - y_mean)**2).sum()

    R2 = SSR / SS0
    return R2
```

```python

print("R1234 =", get_R(['X1','X2','X3','X4']))
print("R123  =", get_R(['X1','X2','X3']))
print("R234  =", get_R(['X2','X3','X4']))
print("R124  =", get_R(['X1','X2','X4']))
print("R134  =", get_R(['X1','X3','X4']))
```

```python
def fisher_stepwise(R_large, R_small, k):
    n = 15
    return (R_large - R_small) * (n - k - 1) / (1 - R_small)
```
