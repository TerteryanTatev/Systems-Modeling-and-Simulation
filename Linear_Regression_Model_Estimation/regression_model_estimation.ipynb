```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'N': [ i for i in range(1,21)],
    'X1': [1, 1, 2, 1, 1, 2, 2, 2, 1, 1, 2, 1, 2, 2, 2, 1, 1, 2, 1, 2],
    'X2': [4, 3, 1, 2, 2, 1, 4, 3, 2, 1, 2, 3, 4, 4, 1, 2, 1, 3, 3, 1],
    'X3': [2, 1, 5, 1, 4, 3, 2, 5, 3, 1, 4, 3, 5, 1, 3, 4, 5, 1, 5, 4],
    'Y':  [50, 22, 65, 38, 58, 34, 54, 95, 40, 18, 60, 52, 100, 45, 36, 61, 59, 32, 89, 88]
})

df
```

```python
x1_mean, x2_mean, x3_mean, y_mean = df['X1'].mean(), df['X2'].mean(), df['X3'].mean(), df['Y'].mean()

df['x1_'] = df['X1'] - x1_mean
df['x2_'] = df['X2'] - x2_mean
df['x3_'] = df['X3'] - x3_mean
df['y_']  = df['Y']  - y_mean


sx1 = np.sqrt(np.sum(df['x1_']**2)/19)
sx2 = np.sqrt(np.sum(df['x2_']**2)/19)
sx3 = np.sqrt(np.sum(df['x3_']**2)/19)
sy  = np.sqrt(np.sum(df['y_']**2)/19)
print("Ստանդարտ շեղումներ:"f"sx1 = {sx1:.4f}, sx2 = {sx2:.4f}, sx3 = {sx3:.4f}, sy = {sy:.4f}\n")
```

```python

def r(a,b):
    return np.sum((a - a.mean())*(b - b.mean())) / np.sqrt(np.sum((a - a.mean())**2)*np.sum((b - b.mean())**2))

r_yx1 = r(df['Y'], df['X1'])
r_yx2 = r(df['Y'], df['X2'])
r_yx3 = r(df['Y'], df['X3'])

r_x1x2 = r(df['X1'], df['X2'])
r_x1x3 = r(df['X1'], df['X3'])
r_x2x3 = r(df['X2'], df['X3'])

print("Կորելյացիաներ՝")
print(f"r(y,x1) = {r_yx1:.4f}")
print(f"r(y,x2) = {r_yx2:.4f}")
print(f"r(y,x3) = {r_yx3:.4f}")
print(f"r(x1,x2) = {r_x1x2:.4f}")
print(f"r(x1,x3) = {r_x1x3:.4f}")
print(f"r(x2,x3) = {r_x2x3:.4f}")
```

```python
R = np.array([
    [1,       r_x1x2,  r_x1x3],
    [r_x1x2,  1,       r_x2x3],
    [r_x1x3,  r_x2x3,  1]
])
R
```

```python
R0 = np.array([
    [r_yx1],
    [r_yx2],
    [r_yx3]
])
R0
```

```python
a = np.linalg.inv(R) @ R0
a
```

```python
Z = df[['X1','X2','X3','Y']].apply(lambda x: (x - x.mean())/x.std())
Z
```

```python
Z_Y = Z['X1']*a[0] + Z['X2']*a[1] + Z['X3']*a[2]
Z_Y
```

```python
Y_hat = df['Y'].mean() + Z_Y * df['Y'].std()
Y_hat
```

```python
df['Y_hat'] = Y_hat
df
```

```python

SSO = np.sum((df["Y"] - y_mean) ** 2)
SSR = np.sum((df["Y_hat"] - y_mean) ** 2)
SSE = np.sum((df["Y"] - df["Y_hat"]) ** 2)
SSO, SSR, SSE
```

```python
R_hat = SSR / SSO
R_hat
```

```python
sigma = SSE / (20-4-1)
msr = SSR / 4
f = msr / sigma
f
```

```python
b1 = a[0] * (sy / sx1)
b2 = a[1] * (sy / sx2)
b3 = a[2] * (sy / sx3)

b0 = y_mean - (b1*x1_mean + b2*x2_mean + b3*x3_mean)
Y_hat2 = b0 + df['X1']*b1 +df['X2']*b2 +df['X3']*b3
df['Y_hat2'] = Y_hat2
df
```
