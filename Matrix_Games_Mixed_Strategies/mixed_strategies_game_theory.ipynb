```python
import numpy as np
import matplotlib.pyplot as plt

a, b, n = 2, 10, 15
r = np.random.rand(n)
x = a + (b - a) * r
print('R =', r)
print('x=',x)
```

```python
lmbda = 2
r = np.random.rand(n)
x = -(1 / lmbda) * np.log(r)
print('R =', r)
print('x=',x)

```
