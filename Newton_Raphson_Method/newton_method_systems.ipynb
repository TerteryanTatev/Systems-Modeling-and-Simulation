```python
import numpy as np


def F(x, y):
    return np.array([
        np.sin(x - 1) + y - 1.3,
        x - np.sin(y + 1) - 0.8
    ])

def J(x, y):
    return np.array([
        [np.cos(x - 1), 1],
        [1, -np.cos(y + 1)]
    ])


eps = 1e-10
n = 50


x, y = 1.0, 1.0

print(x, y)

for k in range(n):
    Fv = F(x, y)
    Jv = J(x, y)


    z = np.linalg.solve(Jv, Fv)


    x_new = x - z[0]
    y_new = y - z[1]

    print(f"Iter {k}: x = {x_new:.5f}, y = {y_new:.5f}")


    if np.linalg.norm(z) < eps:
        break

    x, y = x_new, y_new

print("x =", x_new)
print("y =", y_new)
```

```python

eps = 1e-10
n = 50
x, y = 1.0, 1.0


J0 = J(x, y)


for k in range(1, n):
    Fk = F(x, y)


    z = np.linalg.solve(J0, Fk)

    x_new = x - z[0]
    y_new = y - z[1]

    print(f"Iter {k}: x={x_new:.5f}, y={y_new:.5f}")

    if np.linalg.norm(z) < eps:
        x, y = x_new, y_new
        break

    x, y = x_new, y_new

print("x =", x)
print("y =", y)

```
