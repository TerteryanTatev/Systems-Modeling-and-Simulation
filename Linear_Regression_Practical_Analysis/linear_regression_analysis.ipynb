```python
import numpy as np
```

```python
import pandas as pd
df = pd.DataFrame({
    'n': [i for i in range(1, 21)],
    'x1': [1, 2, 2, 1, 1, 1, 3, 1, 1, 1, 1, 4, 3, 1, 1, 1, 3, 2, 1, 2],
    'x2': [1, 1, 1, 1, 1, 1, 2, 2, 1, 3, 2, 3, 1, 1, 2, 1, 3, 2, 2, 3],
    'x3': [2, 3, 2, 1, 3, 2, 3, 3, 1, 4, 3, 4, 3, 2, 4, 1, 4, 3, 3, 4],
    'y' : [15000, 28000, 26000, 17000, 18000, 14000, 42000, 19000, 13500,
           21000, 22000, 48000, 30000, 18500, 23000, 16000, 37000, 24000,
           17500, 28000]
})

print(df)
```

```python
import numpy as np
x1_mean, x2_mean, x3_mean, y_mean = df['x1'].mean(), df['x2'].mean(), df['x3'].mean(), df['y'].mean()

df['x1_'] = df['x1'] - x1_mean
df['x2_'] = df['x2'] - x2_mean
df['x3_'] = df['x3'] - x3_mean
df['y_']  = df['y']  - y_mean
print(df[['x1_', 'x2_', 'x3_', 'y_']])



sx1 = np.sqrt(np.sum(df['x1_']**2)/(len(df)-1))
sx2 = np.sqrt(np.sum(df['x2_']**2)/(len(df)-1))
sx3 = np.sqrt(np.sum(df['x3_']**2)/(len(df)-1))
sy  = np.sqrt(np.sum(df['y_']**2)/(len(df)-1))
print("Ստանդարտ շեղումներ:"f"sx1 = {sx1:.4f}, sx2 = {sx2:.4f}, sx3 = {sx3:.4f}, sy = {sy:.4f}\n")

def r(a,b):
    return np.sum((a - a.mean())*(b - b.mean())) / np.sqrt(np.sum((a - a.mean())**2)*np.sum((b - b.mean())**2))

r_yx1 = r(df['y'], df['x1'])
r_yx2 = r(df['y'], df['x2'])
r_yx3 = r(df['y'], df['x3'])

r_x1x2 = r(df['x1'], df['x2'])
r_x1x3 = r(df['x1'], df['x3'])
r_x2x3 = r(df['x2'], df['x3'])

print("Կորելյացիաներ՝")
print(f"r(y,x1) = {r_yx1:.4f}")
print(f"r(y,x2) = {r_yx2:.4f}")
print(f"r(y,x3) = {r_yx3:.4f}")
print(f"r(x1,x2) = {r_x1x2:.4f}")
print(f"r(x1,x3) = {r_x1x3:.4f}")
print(f"r(x2,x3) = {r_x2x3:.4f}")
print()


R = np.array([
    [1,        r_x1x2, r_x1x3],
    [r_x1x2,   1,      r_x2x3],
    [r_x1x3,   r_x2x3, 1]
])
print("R=",R)
R0 = np.array([r_yx1, r_yx2, r_yx3])
print("R0",R0)

a = np.linalg.inv(R) @ R0

a1, a2, a3 = a
print("Ստանդարտացված (β) գործակիցներ՝")
print(f"a1 = {a1:.4f}, a2 = {a2:.4f}, a3 = {a3:.4f}")
print()

b1 = a1 * (sy / sx1)
b2 = a2 * (sy / sx2)
b3 = a3 * (sy / sx3)

b0 = y_mean - (b1*x1_mean + b2*x2_mean + b3*x3_mean)

print("Ռեգրեսիոն հավասարման գործակիցներ՝")
print(f"b0 = {b0:.4f}")
print(f"b1 = {b1:.4f}")
print(f"b2 = {b2:.4f}")
print(f"b3 = {b3:.4f}")
print()
print(f"Վերջնական հավասարում՝ ŷ = {b0:.4f} + {b1:.4f}x1 + {b2:.4f}x2 + {b3:.4f}x3")
```

```python
A = np.array([
    [20, df['x1'].sum(), df['x2'].sum(), df['x3'].sum()],
    [df['x1'].sum(), df['x1'] @ df['x1'], df['x1'] @ df['x2'], df['x1'] @ df['x3']],
    [df['x2'].sum(), df['x2'] @ df['x1'], df['x2'] @ df['x2'], df['x2'] @ df['x3']],
    [df['x3'].sum(), df['x3'] @ df['x1'], df['x3'] @ df['x2'], df['x3'] @ df['x3']],
])
print(A)
```

```python
C=np.array([df['y'].sum(),
            df['x1'] @ df['y'],
            df['x2'] @ df['y'],
            df['x3'] @ df['y'],
])
print(C)
```

```python
A_inv = np.linalg.inv(A)
b = A_inv @ C
print("b=",b)
print("A_inv=",A_inv)
```

```python
df['y_hat'] = b[0] + b[1]*df['x1'] + b[2]*df['x2'] + b[3]*df['x3']
df
```
